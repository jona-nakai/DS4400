```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[0, 2],
             [0, 1],
             [1, 0],
             [3, 4],
             [5, 2],
             [5, 3]])

y = np.array([-1, -1, -1, 1, 1, 1]).reshape(-1, 1)
```

```python
def dw(w, X, y):
    return 2*w - X.T @ y
```

```python
lr = 0.01
w = np.array([1, 1]).reshape(-1, 1)

for _ in range(1000):
    w = w - lr * dw(w, X, y)

w
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
x_ = np.linspace(0, 5, 1000)
y_ = (w[1] / w[0]) * x_
plt.plot(x_, y_)
plt.show()
```
